# by dedault, in a box plot, any data plot outside
# $Q_1 - 1.5 IQ, Q_3 + 1.5 IQ$ is considered 'outlier'
# The above range, should contain about 99% of the data.
# So anything outside that range is considered outlier (ASSUMING your data is normally distributed)

```python
import seaborn as sns
```

```python
sns.boxplot(data=census_df['2017 Median Age'])
```

```python
sns.boxplot(data=census_df['2017 Median Age'],
            whis=2)
# whis(default=1.5), whis=2: less outliers
```

```python
sns.boxplot(data=census_df['2017 Median Age'],
            whis=3)
```
