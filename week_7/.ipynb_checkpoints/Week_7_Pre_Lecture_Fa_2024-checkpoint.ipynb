# handlings missing value
### handlings missing values often needs domain knowledge.
### Our first go-to is dropping them, that is not always the best solution.
### **Sometimes** it's best to replace missing values with imputer. Be careful when dropping this

# `dropna()` operates on rows. By default if there is one missing value in a row, it drops the whole row.

```python
# drop missing values
df_housing.dropna()
```

```python
# dictate how to drop rows
df_housing.dropna(how = 'all')
# if all values in a row is missing, drop it
```

```python
df_housing
# dropna() wont change original data
```

```python
df_housing.dropna(how='all', inplace=True)
# this change the original data
```

```python
df_housing
```

```python
df_housing.reset_index(drop=True)
# reset index and will not keep old index
```

```python
# no rows completely missing at this point
msno.matrix(df_housing)
```

```python
# when all values of a volumn missing, often times we drop that column
df_housing.dropna(how='all', axis=1)
# it wont change original data
```

```python
df_housing
```

```python
df_housing.dropna(how='all', axis=1, inplace=True)
df_housing
```

```python
msno.matrix(df_housing)
```

### in order to put firstname and lastname together, we need a unique column in each dataframe.
### `primary key`
### in this example, use 'term'

```python
df_mini_pres_1.merge(df_mini_pres_2, 
                     on = 'Term', 
                     how = 'inner', 
                     validate='1:1')
```

```python
df_mini_pres_1.merge(df_mini_pres_2, 
                     on = 'Term', 
                     how = 'outer', 
                     validate='1:1')
```

```python
df_mini_pres_1.merge(df_mini_pres_2, 
                     on = 'Term', 
                     how = 'left', 
                     validate='1:1')
```

```python
df_mini_pres_1.merge(df_mini_pres_2, 
                     on = 'Term', 
                     how = 'left', 
                     validate='1:1',
                    indicator=True)
# new column '_merge'
```

```python
df_mini_pres_1.merge(df_mini_pres_2, 
                     on = 'Term', 
                     how = 'right', 
                     validate='1:1')
```

```python
df_mini_pres_1.merge(df_mini_pres_2, 
                     on = 'Term', 
                     how = 'right', 
                     validate='1:1',
                     indicator=True )
```
