```python
#17
def age_cate(age):
    if age < 40:
        return 'Young'
    if 40<=age<=59:
        return 'Middle-aged'
    if age >=60:
        return 'Old'

df_Purchase['age_cate'] = df_Purchase['Age'].apply(age_cate)
df_Purchase[['Age','age_cate']].head()
```

```python
#18
df_Purchase['sale_price'] = df_Purchase['Price']*df_Purchase['Quantity']*(1-df_Purchase['Discount'])
df_Purchase[['Price','Quantity','Discount','sale_price']]
```

```python
#19
return_c = df_Purchase['Customer_ID'].value_counts().ge(2).sum()
all_c = df_Purchase['Customer_ID'].nunique()
return_percentage = return_c / all_c
return_percentage
```

```python
df_Purchase.head()
```

```python
#20
df_online = df_Purchase[df_Purchase['Store_Type']=='Online']
df_online.groupby(['Product_Category'])['Lead_Time_Days'].mean()
```

```python
#21
def calculate_discount_frequency(discounts):
    """
    Calculate the frequency of discounts in a given series.
    The frequency is defined as the proportion of non-zero discounts.
    
    It accepts discounts (Series): The 'Discount' column for a specific group.
    
    It returns Proportion of non-zero discounts.
    """
    return (discounts > 0).sum() / len(discounts)

df_fre = df_Purchase.groupby('Region')['Discount'].apply(calculate_discount_frequency)
df_dis_mean = df_Purchase.groupby(['Region'])['Discount'].mean()
display(df_fre)
display(df_dis_mean)
```
